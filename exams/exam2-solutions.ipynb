# Exam 2 Solutions

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as spi
import scipy.linalg as la
%matplotlib inline
```

**QUESTION 1**

```python
def exp_integral(a,b,c,n):
    if a < 0 or b < 0 or c < 0 or n < 0:
        print("Error: Negative parameters")
        return None
    x = np.linspace(0,c,n)
    y = x**a * np.exp(-b * x)
    return spi.trapz(y,x)
```

**QUESTION 2**

```python
def pq_integral(p,q,b):
    if p < 0 or q < 0 or b < 0:
        print("Error: Negative parameters")
        return None
    def f(x):
        return x**p / np.sqrt(1 + x**q)
    approx, error = spi.quad(f,0,b)
    return approx
```

**QUESTION 3**

```python
def sin_cos_fun(a,b,t):
    def f(y,t):
        return np.sin(a*t) + np.cos(b*t**2)
    y = spi.odeint(f,0,t)
    plt.plot(t,y)
    plt.show()
```

**QUESTION 4**

```python
def mathieu(a,q,y0,t):
    
    def f(u,t):
        return [ u[1] , -(a - 2*q*np.cos(2*t))*u[0] ]
    
    u = spi.odeint(f,y0,t)
    plt.plot(t,u[:,0])
    plt.show()
```

**QUESTION 5**

```python
def ode_system(a,b,c,u0,t):
    
    def f(u,t):
        return [ a*(u[1] - u[0]), u[0]*(b - u[2]), u[0]*u[1] - c*u[2] ]
    
    u = spi.odeint(f,u0,t)
    plt.plot(u[:,0],u[:,1])
    plt.show()
```

**QUESTION 6**

```python
def eigvals_squared(A):
    M = np.matrix(A)
    eigvals, eigvecs = la.eig(M * M.T)
    return sum(eigvals.real**2)
```
