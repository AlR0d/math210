## 1. Subplots

We can plot several graphs in the figure using the command `plt.subplot`. The first input is the numerber of rows of subplots, the second input is the number of columns, and the third is which subplot to set as the current subplot. All the `pyplot` commands are applied to the current subplot (there is a *current axes* Python object in the background).

```python
t = np.linspace(-2*np.pi,2*np.pi,100)

# Create a figure with 2 rows and 1 columns and set the current subplot to 1 and plot y = sin(t)
plt.subplot(2,1,1)
y1 = np.sin(t)
plt.plot(t,y1)
plt.xlim([-2*np.pi,2*np.pi])
plt.ylim([-1,1])
plt.title('Since and Cosing Waves')

# Set the current subplot to 2 and plot y = cos(t)
plt.subplot(2,1,2)
y2 = np.cos(t)
plt.plot(t,y2)
plt.xlim([-2*np.pi,2*np.pi])
plt.ylim([-1,1]);
```

Let's plot
$$
y = \sum_{n=0}^N \frac{\sin(2\pi(2n+1)t)}{2n+1}
$$ for $N=0, \dots, 5$ in a 2 by 3 grid of subplots.

```python
plt.figure(figsize=(15,10))
t = np.linspace(0,3,1000)
nrows = 2
ncols = 3
y = 0
for n in range(0,nrows*ncols):
    plt.subplot(nrows,ncols,n+1)
    y = y + np.sin(2*np.pi*(2*n+1)*t)/(2*n+1)
    plt.plot(t,y)
    plt.ylim([-1,1])
    plt.title('Square Wave N = ' + str(n))
```

## 2. Scatter Plots

In matplotlib, we can add color and size dimensions to our scatter plots. For example, let's plot 4 dimensions of random data.

```python
n = 100
samples = np.random.randn(n,2)
colors = np.random.randn(n)
sizes = np.random.randint(50,151,n)
plt.scatter(samples[:,0],samples[:,1],s=sizes,c=colors)
plt.xlim([-4,4])
plt.ylim([-4,4]);
```

## 3. Examples: Plotting data from the web

In NumPy, there is a command to load external data called `np.genfromtxt`. We can load data from a URL. (But beware that the data set will be downloaded and saved to your current directory.)

```python
price_data = np.genfromtxt('https://www.quandl.com/api/v3/datasets/BP/CRUDE_OIL_PRICES.csv',skip_header=1,delimiter=',',usecols=2)
```

```python
price_data[:5]
```

Notice that price_data is yearly price of crude oil in 2014 dollars from 2014 to 1861.

```python
plt.plot(price_data[::-1]);
```
